#### Imports

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from ggplot import *

from sklearn.cluster import KMeans
from sklearn.mixture import GMM, DPGMM
```

#### Load data

```python
total_watt = pd.read_csv('total_watt.csv', parse_dates=[0], names=['time','value'], index_col=0)
```

#### Exploration

```python
total_watt
```

```python
total_watt['time'] = total_watt.index
total_watt['value'].plot(figsize=(15,5))
(ggplot(total_watt, aes(x='time', y='value')) + geom_line())
```

```python
total_watt['day'] = total_watt.index.dayofyear
total_watt['weekday'] = total_watt.index.weekday
total_watt['hour'] = total_watt.index.hour
```

```python
total_watt.groupby('day').aggregate(sum)['value'].plot()
total_watt.boxplot(column='value', by='day')
print (ggplot(total_watt, aes('day', 'value')) +
       geom_point(alpha=0.1,color='blue')
       )
```

```python
total_watt.groupby('weekday').aggregate(sum)['value'].plot()
total_watt.boxplot(column='value', by='weekday')
print (ggplot(total_watt, aes('weekday', 'value')) +
       geom_point(alpha=0.1,color='blue') + 
       scale_x_date(limits=(-0.5,6.5))
       )
```

```python
total_watt.groupby('hour').aggregate(np.mean)['value'].plot()
total_watt.boxplot(column='value', by='hour')
print (ggplot(total_watt, aes('hour', 'value')) +
       geom_point(alpha=0.1,color='blue')
       )
```

#### Clustering

```python
total_watt['rank'] = total_watt['value'].rank()
total_watt.sort(columns='value').plot(y='value', x='rank')
(ggplot(total_watt.sort(columns='value'), aes(x='rank', y='value')) + geom_line())
```

```python
total_watt['value'].plot(kind='kde', xlim=(0,10000))
(ggplot(total_watt, aes(x='value')) + geom_density())
```

```python
kmclust = KMeans(n_clusters=3)
kmclust.fit(total_watt.get(['value']).values)
total_watt['kmclust'] = kmclust.predict(total_watt.get(['value']).values)
(ggplot(total_watt, aes(x='rank', y='value', colour='factor(kmclust)')) + geom_point())
```

```python
gmmclust = GMM(n_components=3)
gmmclust.fit(total_watt.get(['value']).values)
total_watt['gmmclust'] = gmmclust.predict(total_watt.get(['value']).values)
(ggplot(total_watt, aes(x='rank', y='value', colour='factor(gmmclust)')) + geom_point())
```
