Visualization of Energy Consumptions (Cambridge Energy Data Lab - Challenge 2)
----
This notebook will show a series of visualization of the data provided for the Challenge 2.

The following visualizations are proposed:

* 1) A histogram of the distribution of the consumption values;
* 2) A scatterplot of all the values provided;
* 3) A box plot of the consumption values per day;
* 4) A bar chart of the total consumption per day;
* 5) A half hourly box plot of the consumption values;
* 6) A half hourly bar chart of the total consumption.

Each graph comes with the code snippet used to make it, a brief description and some interpretations.

```python
from numpy import loadtxt,array,sort
from numpy import str as numpystr, float as numpyfloat
from pylab import show,plot_date,xticks,figure,legend,bar
from pylab import xlabel,ylabel,hist,vlines,NullLocator
from datetime import datetime
from matplotlib.dates import DateFormatter
```

```python
# import the data from the csv
data = loadtxt('../../data/total_watt.csv', dtype=numpystr,delimiter=',')
get_date = lambda x : datetime.strptime(x,'%Y-%m-%d %H:%M:%S')
timestamp = array([get_date(d) for d in data[:,0]])
consumption = data[:,1].astype(numpyfloat)
```

To begin, we can observe the distribution of the consumption values through the following summary statistics:

```python
# 5 number summary
from numpy import percentile,median,std
print 'minimum:\t',min(consumption)
print '1s quartile:\t',percentile(consumption,25)
print 'median:\t\t',median(consumption)
print '3rd quartile:\t',percentile(consumption,75)
print 'maximum:\t',max(consumption) 
```

and the histrogram of the values:

```python
# histogram of the energy consumption values
figure(figsize=(16, 7))
vlines(percentile(consumption,25),1,400,linestyles='--',colors=(.4,.4,.3),linewidth=3,alpha=.7)
vlines(median(consumption),1,400,linestyles='--',colors=(.7,0,.3),linewidth=3,alpha=.7)
vlines(percentile(consumption,75),1,400,linestyles='--',colors=(.4,.4,.3),linewidth=3,alpha=.7)
hist(consumption,bins=100,alpha=.4)
xlabel('Energy Consumption (W)')
show()
```

From the summary statistcs we can see that the median (red dashed line in the histrogram) and the maximum value are very far, this means that we have a high number of outliers. As expected, the distribution is right skewed and unimodal, which suggests that the outliers correspond to high consumption values and that the regular consumption values are concentrated between the first quartile (dash gray line on the left) and the third quartile (dashed gray line on the rigth).

Now, the data will be clustered in three groups using the kmeans algorithm and the result will be showed with a scatterplot.

```python
# cluster the values into 3 groups: low, medium, and high energy consumption
from scipy.cluster.vq import kmeans,vq
centroids,_ = kmeans(consumption,3)
idx,_ = vq(consumption,sort(centroids)) # assign each sample to a cluster
```

```python
# scatterplot
ax = figure(figsize=(15, 7)).gca()
ax.xaxis.set_major_formatter(DateFormatter('%m/%d'))
plot_date(timestamp[idx==0],consumption[idx==0],'.',color=(.1,.1,.4))
plot_date(timestamp[idx==1],consumption[idx==1],'.',color=(.3,0,.3))
plot_date(timestamp[idx==2],consumption[idx==2],'.',color=(.8,0,.2))
ylabel('Energy Consumption (W)')
xlabel('Timestamp')
legend(['low', 'medium','high'])
plot_date(timestamp,consumption,'-k',alpha=.05)
show()
```

In the graph above we have the time on the x axis and the energy consumption on the y axis. The colors show the three groups found by the kmeans algorithm.  The first group (blue points) contains the consumption values below the mean that can be considered as low consumption values, the second one (purple points) contains the values right above the mean that can be considered medium consumption, the third group (red points) contains the extreme values, which can be considered high. Hovewer, the regular consumption lays between the first two cluesters. (It is to notice that the hole in scatterplot shows that there is missing data for the 4th and 5th of May)

The scatterplot above shows some peaks in the values that suggest a periodicity trought the days of the week. This will be investigated comparing the distribution of the data for each day using a box plot.

```python
# dayly box plot of the energy consumption
from pylab import boxplot
tobox = []
tobox.append([])
oldd = timestamp[0]
daycnt = 0
labels = [timestamp[0].strftime('%a %m/%d')]
for d,v in zip(timestamp,consumption): # arranging the data as required by the boxplot function
    if oldd.day == d.day and oldd.month == d.month:
        tobox[daycnt].append(v)        
    else:
        labels.append(d.strftime('%a %m/%d'))
        daycnt += 1
        tobox.append([])
        oldd = d
figure(figsize=(16, 6))            
boxplot(tobox)
xticks(range(len(labels)),labels,rotation=40)
ylabel('Energy Consumption (W)')
xlabel('Day (each box considers all the samples collected in one day)')
show()
```

In the graph above every box describes the distribution of the consumption values for a single day. The box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data. Flier points are those past the end of the whiskers (those points are considered outliers). 

In this graph we can notice that in a good number of cases the consumptions were, on average, higher during the weekends. Also, during the weekends we can notice the highest number of outliers (very high consumption values). Hoverwer, the periodicity is not clear because for two of weekends considered, the ones located around the days 05/08 and 05/16, we didn't have any high consumptions.

A similar observation can be made looking at the following graph:

```python
tot_consumption = array([sum(dd) for dd in tobox])
centroids,_ = kmeans(tot_consumption,3)
idx,_ = vq(tot_consumption,sort(centroids))
colors = [(.1,.1,.4),(.3,0,.3),(.8,0,.2)]
figure(figsize=(16, 6))            
bar(range(len(tot_consumption)),tot_consumption,color=[colors[i] for i in idx])
xticks(range(len(labels)),labels,rotation=40) # reuse of the xtick
ylabel('Total Energy Consumption in a day (W)')
xlabel('Day')
show()
```

In this chart each bar shows the total consumption per day (sum of all the consumptions of the day). The consumption have been clustered in low (blue), medium (violet) and high (red) consumption. Again, the clustering algorithm applied is kmeans and the coloring provided in this case makes the weekly pattern mentioned above more easy to notice.

Now, to study the consumption during the day a box plot of the values for each 30 minutes will be showed.

```python
# half hour boxplot
tobox = [list() for _ in range(48)]
for h in range(0,24):
    for d,v in zip(timestamp,consumption):
        if d.hour == h:
            if d.minute < 30:
                tobox[2*h].append(v)
            else:
                tobox[2*h+1].append(v)
figure(figsize=(16, 6))            
boxplot(tobox)
labels = []
for h in range(24):
    labels.append('%d:00-%d:30'%(h,h))
    labels.append('%d:30-%d:59'%(h,h))
xticks(range(0,48),labels,rotation=50)
ylabel('Energy Consumption (W)')
xlabel('Half hour')
show()
```

Here, each box describes the distribution fo the data in 30 minutes. It is easy to notice that the average consumption is higher between 10:00 and 13:00. The large amount of outliers between 15:30 and 23:00 suggests that the consumption in the afteernoon and the evening is irregular. Also, there is an increase of the consumption during the evening hours (20:00-1:00).

The same pattern can be observed looking at the total consumption in the 30 minutes:

```python
tot_consumption = array([sum(dd) for dd in tobox])
centroids,_ = kmeans(tot_consumption,3)
idx,_ = vq(tot_consumption,sort(centroids))
colors = [(.1,.1,.4),(.3,0,.3),(.8,0,.2)]
figure(figsize=(16, 6))            
bar(range(len(tot_consumption)),tot_consumption,color=[colors[i] for i in idx])
xticks(range(len(tot_consumption)),labels,rotation=40) # reuse of the xtick
ylabel('Total Energy Consumption in 30 min (W)')
xlabel('Half hour')
show()
```

In this chart, each bar shows the sum of all the consumption values recorded in the interval of 30 minutes across the entire observation period. The values have been clustered as above and the color scheme is still the same.
