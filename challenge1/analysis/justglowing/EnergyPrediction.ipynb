Energy Production Data Simulation (Challenge 1)
----

This notebook contains the code used for the Energy Production Data Simulation challenge.

To solve the proposed problem various regression techniques have been tried. Only the variables Month, Temperature and Daylight have been considered suitable for the prediction and the following workflow have been applied to select the regressor and its parameters:

* Split the training set in train set and development set;
* Tune the paramiter on the train set and test the best regressor on the development set;
* Finally, use the test set to estimate the final result.

The best results have been achieved using Support Vector Regression (SVR) and Elestic Net Regression. The Elastic Net algorithm provided the best scores during the tests and seemed able to generalize the problem avoiding overfitting. The results obtained with SVR were slightly worse and the model was also harder to tune due to recurrent overfitting and the amount of time required by the training process.

```python
from numpy import genfromtxt,mean,abs
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn import preprocessing
```

```python
# import training data
train = genfromtxt('../../data/training_dataset_500.csv',usecols=(4,5,6) ,delimiter=',',skip_header=1)
train = preprocessing.normalize(train,norm='l2')
#train = preprocessing.scale(train)
ytrain = genfromtxt('../../data/training_dataset_500.csv',usecols=(7) ,delimiter=',',skip_header=1)
```

```python
# import test data
test = genfromtxt('../../data/test_dataset_500.csv',usecols=(4,5,6) ,delimiter=',',skip_header=1)
test = preprocessing.normalize(test,norm='l2')
#test = preprocessing.scale(test)
ytest = genfromtxt('../../data/test_dataset_500.csv',usecols=(7) ,delimiter=',',skip_header=1)
```

```python
figure(figsize=(15,4))
for i in range(train.shape[1]):
    subplot(1,train.shape[1],i+1)
    plot(train[:,i],ytrain,'.')
```

```python
def mape(regressor, x, y_true):
    # Mean Absolute Precentage Error    
    # IMPORTANT: it's negative to pick automatically the min value during the Grid Search
    y_pred = regressor.predict(x)
    return -mean(abs((y_true - y_pred) / y_true)) # notice, not multiplied by 100
```

```python
def show_grid_results(clf,dev,ydev):
    print "Best parameters set found on development set:"
    print ' ',clf.best_estimator_
    print "Grid scores:"
    for params, mean_score, scores in clf.grid_scores_:
        print "   %0.5f (+/-%0.03f) for %r" % (-mean_score, scores.std() / 2, params)
    print "Results on the development set:"
    print ' ',-mape(clf.best_estimator_,dev,ydev)
```

```python
# splitting the testset in test and dev set
X_train, X_testsplit, y_train, y_testsplit = train_test_split(train, ytrain, test_size=0.4)
print 'trainset size:',X_train.shape[0]
print 'testset size:',X_testsplit.shape[0]
```

```python
# tuning and testing of Support Vector Regessor
from sklearn.svm import SVR
tuned_parameters = [{'kernel': ['rbf',],'C':[1e-5,1,10,100],'gamma':[1e-5,1,10,100]}]
clf = GridSearchCV(SVR(), tuned_parameters, cv=3, scoring=mape)
clf.fit(X_train,y_train)
show_grid_results(clf,X_testsplit,y_testsplit)
print 'Result on the test set:'
clf.best_estimator_.fit(train,ytrain) # train on the whole trainset
print '*** Result on the test set:\n  ',-mape(clf.best_estimator_,test,ytest)
```

SVR seems attracted by high values of $C$ and $\gamma$ but it's able to achive better results only on the dev set. Which means that the parameters are overfitted. NOTE: the parameters reported here are only the last ones tried during the selection process.

```python
# tuning and testing of Elastic Net Regression
from sklearn.linear_model import ElasticNet
tuned_parameters = [{'alpha': [1e-5,1e-3,.2,.5,1.0],'l1_ratio':[1e-3,.1,0.5,.8,1.0]}]
clf = GridSearchCV(ElasticNet(), tuned_parameters, cv=20, scoring=mape)
clf.fit(X_train,y_train)
show_grid_results(clf,X_testsplit,y_testsplit)
clf.best_estimator_.fit(train,ytrain) # train on the whole trainset
print '*** Result on the test set:\n  ',-mape(clf.best_estimator_,test,ytest)
```

```python
# writing the csv for the submission
with open('predicted_energy_production.csv','wb') as outfile:
    outfile.write('"House","EnergyProduction"\n')
    for i,p in enumerate(clf.best_estimator_.predict(test)):
        outfile.write('%d, %f\n'%(i+1,p))

# writing the mape value
with open('mape.txt','wb') as outfile:
    outfile.write('MAPE: %f\n'%-mape(clf.best_estimator_,test,ytest))
```
