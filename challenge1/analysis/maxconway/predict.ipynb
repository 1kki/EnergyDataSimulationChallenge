## Prediction

Time to dive into some prediction. We haven't much data, so we'll start with a linear model.

```python
res = smf.glm('EnergyProduction ~ Temperature + Daylight + abs(Month-6) + C(House)', training).fit()
```

```python
np.mean(np.abs(res.predict(training) - training['EnergyProduction']) / training['EnergyProduction'])
```
