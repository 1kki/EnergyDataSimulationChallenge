# CamEnergy Data Lab: Challenge 1 - Energy Production Data Simulation

The challenge here is to predict the PV Energy production of 500 houses based on historic data (Temperature and Daylight). Here's a data sample:

```python
#To display plots in the same window
%pylab inline  
```

```python
import numpy as np
import pylab as pl
from sklearn import feature_selection
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import svm
from sklearn import neighbors
from sklearn import ensemble
from sklearn import grid_search
from sklearn import metrics
from utility_extra import mape
```

Load data

```python
# 0-ID  1-Label  2-House  3-Year  4-Month  5-Temperature  6-Daylight  7-EnergyProduction
data_train = np.loadtxt("../../data/training_dataset_500.csv", delimiter=',', skiprows=1)
data_test = np.loadtxt("../../data/test_dataset_500.csv", delimiter=',', skiprows=1)
x_train = data_train[:, :7]
y_train = data_train[:, 7]
x_test = data_test[:, :7]
y_test = data_test[:, 7]
```

Data preprocessing, I scaled and normalized data, but only scaling helped the fit.

```python
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

#normalizer = preprocessing.Normalizer()
#x_train = normalizer.fit_transform(x_train)
#x_test = normalizer.transform(x_test)
```

Intuitively it seems like the most important parameters/dimensions should be: House, Month, Temperature and Daylight. But to be sure (maybe the data were tricked??) I did a univariate feature selection. I did a SVM regression and looked at the weight of the individual parameters.  

```python
# Univariate feature selection with F-test for feature scoring
selector = feature_selection.SelectPercentile(feature_selection.f_regression, percentile=60)
selector.fit(x_train, y_train)

clf = svm.SVR(kernel='linear')
clf.fit(x_train, y_train)
svm_weights = (clf.coef_ ** 2).sum(axis=0)
svm_weights /= svm_weights.max()

clf_selected = svm.SVR(kernel='linear')
clf_selected.fit(selector.transform(x_train), y_train)
svm_weights_selected = (clf_selected.coef_ ** 2).sum(axis=0)
svm_weights_selected /= svm_weights_selected.max()
```

```python
print "MAPE score linear SVR: ", mape( y_test, clf.predict(x_test))
print "MAPE score linear SVR with selector: ", mape( y_test, clf_selected.predict(selector.transform(x_test)))
```

Plot results:

```python
X_indices = np.arange(x_train.shape[-1])
pl.bar(X_indices - .25, svm_weights, width=.2, label='SVR weight', color='r')
pl.bar(X_indices[selector.get_support()] - .05, svm_weights_selected, width=.2,
       label='SVR weights after selection', color='b')
pl.title("Comparing feature selection")
pl.xlabel('Feature number')
pl.xticks(X_indices,['ID',  'Label', 'House', 'Year', 'Month', 'Temperature', 'Daylight'])
pl.axis('tight')
pl.legend(loc='upper left')
pl.show()
```

Something seems odd, the 'House' parameter has no weight at all, but we would expect wide variations from home to home. Let's look at the data for the relevent parameters and do a linear regression to get a sense of what's going on:

```python
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
pred = regr.predict(x_test)
print "MAPE: ", mape(y_test, regr.predict(x_test)), "%"
```

```python
xlabel = ['House','Month','Temperature','Daylight']
data_col = [2, 4, 5, 6]
for i, icol in enumerate(data_col):
    plt.subplot(2, 2, i + 1)
    plt.scatter(x_train[:, icol], y_train, s=1)
    xr = np.asarray([min(x_train[:, icol]), max(x_train[:, icol])])
    yr = xr * regr.coef_[icol] + regr.intercept_
    plt.plot(xr, yr, color = 'red')
    pl.axis('tight')
    plt.title(xlabel[i])
    plt.ylabel('E')
    subplots_adjust(hspace=.35)
plt.show()
```

We can see the Energy production in function of the individual parameters with the fitted slope in red. Temperature and Daylight are fitted fine, and clearly Month is not linear. But the real problem here is with Houses, they aren't sorted in any particular order, so no trend is found in the data. I tried to sort the houses with their averaged Energy production and that improved the fit a lot. But sorting is still just an approximation, so I added a parameter to the data "Mean Energy Production" which is the average energy production of that house. Then I substracted this value to the Energy production to get just the variance. That improved the fit quite a bit as well.  

```python
# Columns of data:
# 0-ID  1-Label  2-House  3-Year  4-Month  5-Temperature  6-Daylight  7-MeanEnergyProduction 8-DiffE
data_train = np.loadtxt("../../data/training_dataset_500_nh.csv", delimiter=',')
data_test = np.loadtxt("../../data/test_dataset_500_nh.csv", delimiter=',')
x_train = data_train[:, [4, 5, 6, 7]]
y_train = data_train[:,8]
x_test = data_test[:, [4, 5, 6, 7]]
y_test = data_test[:, 8]
scaler = preprocessing.StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Then I tested a few models:

```python
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
pred = regr.predict(x_test)
print "Lin reg"
print "MAPE: ", mape(y_test, regr.predict(x_test))

ridge = linear_model.Ridge(alpha = 0.5)
ridge.fit(x_train, y_train)
print "Ridge"
print "MAPE: ", mape(y_test, ridge.predict(x_test))

lasso = linear_model.Lasso(alpha = 0.1)
lasso.fit(x_train, y_train)
print "Lasso"
print "MAPE: ", mape(y_test, lasso.predict(x_test))

lassolars = linear_model.LassoLars(alpha = 0.1)
lassolars.fit(x_train, y_train)
print "LAssoLars"
print "MAPE: ", mape(y_test, lassolars.predict(x_test))

lassocv = linear_model.LassoCV()
lassocv.fit(x_train, y_train)
print "LassoCV"
print "MAPE: ", mape(y_test, lassocv.predict(x_test))

lassolarscv = linear_model.LassoLarsCV()
lassolarscv.fit(x_train, y_train)
print "LassoLarsCV"
print "MAPE: ", mape(y_test, lassolarscv.predict(x_test))

alpha =0.1
enet = linear_model.ElasticNet(alpha=alpha)
enet.fit(x_train, y_train)
print "ElasticNet"
print "MAPE: ", mape( y_test, enet.predict(x_test))


neigh = neighbors.KNeighborsRegressor(n_neighbors=5)
neigh.fit(x_train, y_train) 
print "KN n=5"
print "MAPE: ", mape( y_test, neigh.predict(x_test))
neigh = neighbors.KNeighborsRegressor(n_neighbors=10)
neigh.fit(x_train, y_train) 
print "KN n=10"
print "MAPE: ", mape( y_test, neigh.predict(x_test))
neigh = neighbors.KNeighborsRegressor(n_neighbors=100)
neigh.fit(x_train, y_train) 
print "KN n=100"
print "MAPE: ", mape( y_test, neigh.predict(x_test))

svr = svm.SVR(kernel='linear')
svr.fit(x_train, y_train)
print "SVR lin"
print "MAPE: ", mape( y_test, svr.predict(x_test))

svr = svm.SVR(kernel='poly', degree=3)
svr.fit(x_train, y_train)
print "SVR poly 3"
print "MAPE: ", mape( y_test, svr.predict(x_test))

svr = svm.SVR(kernel='rbf')
svr.fit(x_train, y_train)
print "SVR RBF"
print "MAPE: ", mape( y_test, svr.predict(x_test))

sgd = linear_model.SGDRegressor()
sgd.fit(x_train, y_train)
print "SGD"
print "MAPE: ", mape( y_test, sgd.predict(x_test))

rfr = ensemble.RandomForestRegressor()
rfr.fit(x_train, y_train)
print "RFR"
print "MAPE: ", mape( y_test, rfr.predict(x_test))
```

That's not bad, but I tried to do a grid search to see if I could improve the fit with better fitting parameters for the SVR and RFR models. The best results were with Random Forest Regression.

```python
rfr = ensemble.RandomForestRegressor(n_estimators=300, max_features='sqrt', min_samples_split=10, min_samples_leaf=2)
rfr.fit(x_train, y_train)
print "RFR"
print "MAPE: ", mape( y_test, rfr.predict(x_test))
```

That seems pretty good, let's now see what the predictions looks like:

```python
y_pred = rfr.predict(x_test)
y_pred = y_pred * data_test[:, 7]
y_test = y_test * data_test[:, 7]
```

```python
pl.scatter(y_test, y_pred, cmap=pl.cm.Paired)
low = min(np.append(y_test, y_pred))
hig = max(np.append(y_test, y_pred))
pl.plot([low,hig], [low,hig], '-')
pl.xlabel('True Energy Produced')
pl.ylabel('Predicted Energy Produced')

pl.show()
```

```python
pl.hist(abs(y_test - y_pred))
pl.xlabel('Error')
pl.ylabel('N')
pl.title('Histogram of errors')
```

```python
print 'Root mean square error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

So the prediction seems pretty good, except for 12 houses where the prediction is quite different. Perhaps a closer look at those specific houses is required.
